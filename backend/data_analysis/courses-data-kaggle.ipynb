Trataremos cada Duration diferente en función del

```python
def extract_num_word(text):
    if not isinstance(text, str):  # maneja NaN o valores no string
        return []
    
    tokens = text.lower().split()
    result = []
    
    for i in range(len(tokens) - 1):
        if tokens[i].isdigit():
            result.append(f"{tokens[i]} {tokens[i+1]}")
    
    return result

df_raw_data['Duration_pairs'] = df_raw_data['Duration'].str.replace(r'\s+', ' ', regex=True).apply(extract_num_word)

print(df_raw_data['Duration_pairs'])
```

```python
print(df_raw_data.loc[df_raw_data['Duration_pairs'].apply(lambda x: len(x)>1), 'Duration_pairs'])
```

```python
def duration_1_len(string):
    if not isinstance(string, str):
        return None
    
    tokens = string.split()
    if len(tokens) < 2:
        return None

    value = float(tokens[0])
    unit = tokens[1]

    if "min" in unit:
        return value * 60
    elif "hour" in unit:
        return value * 3600
    elif "week" in unit:
        return value * 3600 * 10   # supones 10h/semana
    elif "month" in unit:
        return value * 3600 * 10 * 4
    else:
        return None


def duration_2_len(array_s):
    # array_s tipo ["3 months", "5 months"]
    if len(array_s) != 2:
        return None

    dur1 = duration_1_len(array_s[0])
    dur2 = duration_1_len(array_s[1])

    if dur1 is None or dur2 is None:
        return None

    # promedio del rango
    return (dur1 + dur2) / 2


def duration_to_seconds(array_s):
    if not isinstance(array_s, list) or len(array_s) == 0:
        return None
    
    if len(array_s) == 1:
        return duration_1_len(array_s[0])
    
    elif len(array_s) == 2:
        return duration_2_len(array_s)
    
    else:
        return None


df_raw_data["Duration_seconds"] = df_raw_data["Duration_pairs"].apply(duration_to_seconds)
df_data["duration_seconds"] = df_raw_data["Duration_seconds"]
df_data["duration_seconds"].head()
```

Como podemos observar, la cantidad de datos nulos en las distintas columnas del dataframe es muy alta.
A al hora de aplicar técnicas de machine learning a un conjunto de datos, es importante que no haya datos nulos, ya que dificulta la obtención de un modelo correcto.
Tendremos que imputar las columnas con datos vacios. Para cada columna seguiremos una estrategia distinta:
1. Category: añadiremos categoria "non defined" para valores nulos
2. Language: añadiremos categoria "non defined" para valores nulos
3. Course Type: añadiremos categoria "non defined" para valores nulos
4. Sub-Category: añadiremos categoria "non defined" para valores nulos
5. Intro: añadiremos una cadena vacia "" para valores nulos
6. duration_seconds: Le pondremos la mediana de duración de cursos a los valores nulos
7. Ratings_float: Le pondremos 2.5 a los valores nulos

```python
df_data["Sub-Category"].unique()
```

```python
cols = ["Category", "Language", "Course Type", "Sub-Category"]

for col in cols:
    df_data[col] = df_data[col].fillna("Non defined")
df_data["Intro"] = df_data["Intro"].fillna("")
df_data["Ratings_float"] = df_data["Ratings_float"].fillna(2.5)
df_data["duration_seconds"] = df_data["duration_seconds"].fillna(df_data["duration_seconds"].median())
```
